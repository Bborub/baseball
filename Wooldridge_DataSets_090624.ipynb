```python
 import wooldridge as woo
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
```

```python
 affairs = woo.dataWoo('affairs')
```

```python
affairs.head()
```

```python
 age_mean = np.mean(affairs['age'])
 print(age_mean)
```

```python
age_median = np.median(affairs['age'])
print(age_median)
```

```python
age_std = np.std(affairs['age'])
print(age_std)
```

```python
plt.hist(affairs['age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
plt.hist(affairs['yrsmarr'])
plt.xlabel('Years Married')
plt.ylabel('Frequency')
plt.title('Histogram of Years Married')
plt.show()
```
